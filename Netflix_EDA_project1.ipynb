Now we need to categorise vote average column so we can differentiate movies by popuar when it is having popular , average , below_avg, low_popular to describe it more clearly

```python
#convert Vote_Average column to int values present they are object
df['Vote_Average'] = df['Vote_Average'].astype(float)
#creating funtuion to categorise it
def categorize_col(df, col, labels ):
  """
  categorize a certain column based on its quartiles
    ARGS:
    df - dataframe
    col - to be categorised columns names
    labels - list of labels from min to max

    return - df with categorised columns
     """

     #setting edges to cut the columns accordingly
  edges = [df[col].describe()['min'],
              df[col].describe()['25%'],
              df[col].describe()['50%'],
              df[col].describe()['75%'],
              df[col].describe()['max']]

  df[col] = pd.cut(df[col], edges, labels = labels, duplicates='drop')
  return df
```

```python
df.info()
#define labels for edges
labels = ['not_popular', 'below_avg', 'average', 'popular']

cols = 'Vote_Average'
#categorise column based on labels and edges
categorize_col(df, cols, labels)
```
